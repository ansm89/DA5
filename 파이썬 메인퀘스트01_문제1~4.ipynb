```python
# 문제 3번

import random

class Account:

    account_list = []

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.bank_name = "SC은행"
        self.account_number = self.random_number()
        Account.account_list.append(self)

    def random_number(self):
        while True:
            p1 = random.randint(0, 999)
            p2 = random.randint(0, 99)
            p3 = random.randint(0,999999)
            random_account_number = f"{p1:03d}-{p2:02d}-{p3:06d}"

            return random_account_number
                
    def open_Account(self):
          print(f"----- 계좌가 개설되었습니다.-----")
          print(f"은행명 : {self.bank_name}")
          print(f"예금주 : {self.owner}")
          print(f"계좌번호 : {self.account_number}")
          print(f"초기잔액 : {self.balance}원")

    @classmethod
    def get_account_num(cls):
        print(f"개설된 총 계좌 수 : {len(cls.account_list)}개")
        
        return len(cls.account_list)
    


# 실행 코드
acc1 = Account("진우", 500)
acc2 = Account("소라", 0)
acc3 = Account("형준", 10000)


# get_account_num()
Account.get_account_num()
```

```python
# 문제 4번

import random

class Account:

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.bank_name = "SC은행"
        self.account_number = self.random_number()

    def random_number(self):
        while True:
            p1 = random.randint(0, 999)
            p2 = random.randint(0, 99)
            p3 = random.randint(0,999999)
            random_account_number = f"{p1:03d}-{p2:02d}-{p3:06d}"

            return random_account_number
                
    def open_Account(self):
          print(f"----- 계좌가 개설되었습니다. -----")
          print(f"은행명 : {self.bank_name}")
          print(f"예금주 : {self.owner}")
          print(f"계좌번호 : {self.account_number}")
          print(f"초기잔액 : {self.balance}원")

    def deposit(self, amount):
      if amount >= 1:
        self.balance += amount

        print(f"{amount} 원이 입금되었습니다. 잔액은 {self.balance} 원 입니다.")
        
      else:
        print("입금의 경우 1원 이상 입금해주셔야 합니다.")


# 실행 코드
acc1 = Account("진우", 500)


# 입금 메서드
acc1.deposit(100)
acc1.deposit(0)
```
