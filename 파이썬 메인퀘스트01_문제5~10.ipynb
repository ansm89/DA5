```python
# 문제 7번

import random

class Account:

    account_list = []

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.bank_name = "SC은행"
        self.account_number = self.random_number()
        self.deposit_count = 0
        Account.account_list.append(self)

    def random_number(self):
        while True:
            p1 = random.randint(0, 999)
            p2 = random.randint(0, 99)
            p3 = random.randint(0, 999999)
            random_account_number = f"{p1:03d}-{p2:02d}-{p3:06d}"
            return random_account_number
        
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_count += 1
            print(f"{amount:,} 원 입금되었습니다. (입금 횟수 : {self.deposit_count} 회)")

            if self.deposit_count % 5 == 0:
                bonus = int(self.balance * 0.01)
                self.balance += bonus
                print()
                print(f"축하합니다. 5회 입금 이자{bonus:,} 원이 추가되었습니다.")

    def bonus_info(self):
        print(f"----- 계좌 정보 -----")
        print(f"은행명 : {self.bank_name}")
        print(f"예금주 : {self.owner}")
        print(f"계좌번호 : {self.account_number}")
        print(f"잔액 : {self.balance:,} 원")


    
# 실행 코드
acc1 = Account("진우", 1000)


# 5회 입금
acc1.deposit(100)
acc1.deposit(300)
acc1.deposit(800)
acc1.deposit(500)
acc1.deposit(1000)

print()

# 이자 출력
acc1.bonus_info()
print()
```

```python
# 문제 10번

import random

class Account:
    account_list = []
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.bank_name = "SC은행"
        self.account_number = self.random_number()

        self.deposit_count = []
        self.withdraw_count = []

        Account.account_list.append(self)

    def random_number(self):
            while True:
                p1 = random.randint(0, 999)
                p2 = random.randint(0, 99)
                p3 = random.randint(0, 999999)
                random_account_number = f"{p1:03d}-{p2:02d}-{p3:06d}"

                return random_account_number
            
    def open_Account(self):    
          print(f"----- 계좌가 개설되었습니다. -----")
          print(f"은행명 : {self.bank_name}")
          print(f"예금주 : {self.owner}")
          print(f"계좌번호 : {self.account_number}")
          print(f"잔액 : {self.balance:,} 원")
          print(f"-------------------------------")

    def deposit(self, amount):
         if amount >= 1:
              self.balance += amount
              self.deposit_count.append(amount)
              print(f"{self.owner}님 계좌에 {amount:,} 원이 입금되었습니다.")

    def withdraw(self, amount):
            if amount > self.balance:
                print(f"계좌 잔액이 부족합니다. (현재 총 잔액은 {self.balance:,} 원 입니다.)")
            else:
                self.balance -= amount
                self.withdraw_count.append(amount)
                print(f"{self.owner}님 계좌에서 {amount:,} 원이 출금되었습니다.")

    def deposit_history(self):
        print(f"----- {self.owner}님의 입금 내역입니다. -----")
        for amount in self.deposit_count:
            print(f"{amount:,} 원")
            print("---------------------------------------")


    def withdraw_history(self):
        print(f"----- {self.owner}님의 출금 내역입니다. -----")
        for amount in self.withdraw_count:
            print(f"{amount:,} 원")
            print(f"---------------------------------------")

# 실행
acc1 = Account("진우", 10500)


#입금 및 출금
acc1.deposit(3000)
acc1.deposit(25000)
acc1.withdraw(1500)
acc1.withdraw(2300)

print()


# 입출금 내역 출력
acc1.deposit_history()

print()

acc1.withdraw_history()
```
